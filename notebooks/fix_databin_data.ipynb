```python
def list_files(file_dir, file_type=None,get_times=None):
    
    from os import listdir
    from os.path import isfile, join
    
#     try:

    if file_type==None:
        onlyfiles = sorted([ f for f in listdir(file_dir) if isfile(join(file_dir,f)) ])
    else:
        allfiles = sorted([ f for f in listdir(file_dir) if isfile(join(file_dir,f)) ])
        onlyfiles=[]
        for f2 in allfiles:
            try:
                ff=str(f2.lower()).split('.')
                fname=ff[0]
                ftype=ff[1]
                if ftype==file_type.lower():
                    onlyfiles.append(f2)
            except:
                pass 
    if len(onlyfiles)==0:
        print '-->No files found in this directory.'
        return None
    elif get_times!=None:
        file_times={}
        import os.path, time
        for file_name in onlyfiles:
            file_path=file_dir+'\\'+file_name

            last_modified=time.ctime(os.path.getmtime(file_path))
            time_created=time.ctime(os.path.getctime(file_path))
            file_times[file_name]={'time created':time_created,'last modified':last_modified}



        return file_times

    else:
        return onlyfiles
        
#     except Exception, e:
#         print 'Error raised: ',e
#         return None
    
    
                
        
```
